```python
import brightway2 as bw
import perdu as p
```

```python
bw.projects.set_current("US EPA")
```

```python
for _, a in zip(range(50), bw.Database("US LCI")):
    candidates = p.search_naics(a['name'])
    print(a['name'])
    if candidates:
        try:
            for x in range(5):
                print(candidates[x]['score'], candidates[x]['code'], candidates[x]['name'])
        except:
            pass
    else:
        print("Skipped")
```

```python
def get_isic_classification(lst):
    for elem in lst:
        if elem[0] == 'ISIC rev.4 ecoinvent':
            return elem[1]
    return None
```

```python
for _, a in zip(range(50), bw.Database("ecoinvent 3.5 cutoff")):
    candidates = p.search_naics(a['name'])
    print(a['name'])
    print(get_isic_classification(a['classifications']))
    if candidates:
        print(candidates[0]['code'], candidates[0]['name'])
    else:
        print("Skipped")
```
